```python
# 1. f(x) = sin(x)^2 - cos(x)^2
# 1.1. Определить корни.
# 1.2. Найти интервалы, на которых функция возрастает.
# 1.3. Найти интервалы, на которых функция убывает.
# 1.4. Построить график.
# 1.5. Вычислить вершину.
# 1.6. Определить промежутки, на котором f > 0.
# 1.7. Определить промежутки, на котором f < 0.
```

```python
import sympy as sm
import scipy as sc
import math as mt
import pandas as pd
```

```python
x = sm.Symbol('x')
f = sm.sin(x)**2 - sm.cos(x)**2
```

```python
          # 1.1. Определение корней и нахождение производной.
roots = sm.solve(f,x)
print(roots)
for i in roots:
    print(round(i.evalf(),6),"радиан или", mt.degrees(i.evalf()), "градусов.")

derivative = sm.diff(f,x)
print("Производная:", derivative)
```

```python
          # 1.2. Интервалы, на которых функция возрастает.
print(sm.solveset(derivative > 0, x, domain=sm.S.Reals))
print()
sm.pprint(sm.solveset(derivative > 0, x, domain=sm.S.Reals), use_unicode=True)
```

```python
          # 1.3. Интервалы, на которых функция убывает.
print(sm.solveset(derivative < 0, x, domain=sm.S.Reals))
print()
sm.pprint(sm.solveset(derivative < 0, x, domain=sm.S.Reals), use_unicode=True)
```

```python
          # 1.4. График функции.
p1 = sm.plot(f, show=False)
p1.show()
```

```python
          # 1.5. Определение вершин.
print("Вершины:")
sm.solve(derivative,x)  
```

```python
          # 1.6. Определение промежутков, на которых f > 0.
sm.pprint(sm.solveset(f > 0, x, domain=sm.S.Reals), use_unicode=True)
```

```python
          # 1.6. Определение промежутков, на которых f < 0.
sm.pprint(sm.solveset(f < 0, x, domain=sm.S.Reals), use_unicode=True)
```
